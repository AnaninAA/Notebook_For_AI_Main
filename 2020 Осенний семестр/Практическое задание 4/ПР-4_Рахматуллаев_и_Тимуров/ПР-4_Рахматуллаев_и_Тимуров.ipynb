```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```
# Набор данных взят с https://www.kaggle.com/aashi20/top-50-spotify-songs
# Top-50 песен в Spotify в 2019 году.
data = pd.read_csv("top50.csv")
print(data)
```

```
new_data = pd.DataFrame()
```

```
d = pd.get_dummies(data['Genre'])
new_data = pd.concat([new_data, d])
new_data.head()
```

```
d = data['Beats.Per.Minute']
plt.hist(np.log(d), bins=50)
plt.show()
d = np.array(np.log(d)).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Beats.Per.Minute'] = d
new_data.head(10)
```

```
d = data['Energy']
plt.hist(np.log(d), bins=50)
plt.show()
d = np.array(np.log(d)).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Energy'] = d
new_data.head(10)
```

```
d = data['Danceability']
d = np.clip(d, 55, 100)
plt.hist(d**0.5, bins=50)
plt.show()
d = np.array(d**0.5).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Danceability'] = d
new_data.head(10)
```

```
d = data['Loudness..dB..']
d = np.clip(d, -10, 0)
d = np.array(d).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Loudness..dB..'] = d
new_data.head(10)
```

```
d = data['Liveness']
d = np.clip(d, 0, 25)
plt.hist(np.log(d), bins=50)
plt.show()
d = np.array(np.log(d)).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Liveness'] = d
new_data.head(10)
```

```
d = data['Valence.']
plt.hist(d**0.5, bins=50)
plt.show()
d = np.array(d**0.5).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Valence.'] = d
new_data.head(10)
```

```
d = data['Length.']
d = np.clip(d, 140, 500)
plt.hist(d**0.5, bins=50)
plt.show()
d = np.array(d**0.5).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Length.'] = d
new_data.head(10)
```

```
d = data['Acousticness..']
plt.hist(d**0.5, bins=50)
plt.show()
d = np.array(d**0.5).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Acousticness..'] = d
new_data.head(10)
```

```
d = data['Speechiness.']
d = np.clip(d, 0, 40)
plt.hist(d**0.5, bins=50)
plt.show()
d = np.array(d**0.5).reshape(-1, 1)
d = MinMaxScaler().fit_transform(d).flatten()
new_data['Speechiness.'] = d
new_data.head(10)
```
