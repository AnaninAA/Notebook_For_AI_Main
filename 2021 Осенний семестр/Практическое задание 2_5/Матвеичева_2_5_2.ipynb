```python
x = 10
y = 25

sum = x+y
dif = x-y
com = x*y
priv = x/y

print('Сумма равна: ' + str(sum))
print('Разность равна: ' + str(dif))
print('Произведение равно: ' + str(com))
print('Частное равно: ' + str(priv))

expression = ((x**2 + y)*(y -1))/(x-y+y**2)

print('Выражение равно: '+ str(expression))
```
