```python
import numpy as np
import matplotlib.pyplot
```

```python
#Генерация уникального seed
code = "Osmanov"
limit = 2 ** 32
seed = int.from_bytes(code.encode(), "little") % limit

np.random.seed(seed)

#Создание случайную нормально распределенную выборку first_sample и её отсортированную версию second_sample
first_sample = np.random.normal(0, 1, 10000)
second_sample = np.sort(first_sample)

matplotlib.pyplot.hist(first_sample, bins=100)
matplotlib.pyplot.show()
matplotlib.pyplot.hist(second_sample, bins=100)
matplotlib.pyplot.show()
```

```python
#Подвыборка fSample из первых 60% элементов first_sample и вывод
fSample = first_sample[0:6000]
matplotlib.pyplot.hist(fSample, bins=100)
matplotlib.pyplot.show()
```

```python
#Подвыборка sRandSample из 60% случайных элементов из first_sample и вывод
sRandSample1 = np.random.choice(first_sample, 6000, replace=False)
matplotlib.pyplot.hist(sRandSample1, bins=100)
matplotlib.pyplot.show()
```

```python
#Подвыборка sSample, состоящую из первых 60% элементов second_sample и вывод
sSample = second_sample[0:6000]

matplotlib.pyplot.hist(sSample, bins=100)
matplotlib.pyplot.show()
```

```python
#Подвыборка sRandSample2, выбрав в нее 60% случайных элементов из second_sample и вывод
sRandSample2 = np.random.choice(second_sample, 6000, replace=False)
matplotlib.pyplot.hist(sRandSample2, bins=100)
matplotlib.pyplot.show()
```
