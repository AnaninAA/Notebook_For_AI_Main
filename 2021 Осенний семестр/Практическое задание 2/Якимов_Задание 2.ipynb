```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
```

```python
mycode = "Yakimov"
seed_limit = 2 ** 32
my_seed = int.from_bytes(mycode.encode(), "little") % seed_limit

np.random.seed(my_seed)
sample1 = np.random.normal(0, 1, 10000)
sample2 = np.sort(sample1)
# случайная нормально распределенная выборка sample1 
plt.hist(sample1, bins=100)
plt.show()
```

```python
#её отсортированная версия sample2
plt.hist(sample2, bins=100)
plt.show()
```

```python
#подвыборка s_1, состоящая из первых 60% элементов sample1
s_1 = sample1[0:6000]
plt.hist(s_1, bins=100)
plt.show()
```

```python
#подвыборка s_rnd_1, из 60% случайных элементов sample1
s_rnd_1 = np.random.choice(sample1, 6000, replace=False)
plt.hist(s_rnd_1, bins=100)
plt.show()
```

```python
# Выведите и сравните гистограммы для s_1 и s_rnd_1:
# Опишите и объясните результат
#Вывод гистограммы для s_1
plt.hist(s_1, bins=100)
plt.show()
```

```python
#Вывод гистограммы для s_rnd_1
plt.hist(s_rnd_1, bins=100)
plt.show()
```

```python
"""выборка 60% первых элементов sample1 похожа на выборку 60% случайных элементов sample1,
так как sample1 имеет нормально распределена"""

s_2 = sample2[0:6000]
#подвыборка s_2, состоящая из первых 60% элементов sample2
plt.hist(s_2, bins=100)
plt.show()
```

```python
#подвыборка s_rnd_2, состоящая из 60% случайных элементов sample2
s_rnd_2 = np.random.choice(sample2, 6000, replace=False)
plt.hist(s_rnd_2, bins=100)
plt.show()
```

```python
#  Выведите и сравните гистограммы для s_2 и s_rnd_2:
#  Опишите и объясните результат
plt.hist(s_2, bins=100)
plt.show()
```
